```python
import matplotlib.pyplot as plt

# Data for visualizations
data = {
    "Sleep Patterns": {"Pre-COVID": 6.5, "Post-COVID": 8.5},
    "Fast Food Consumption": {"Pre-COVID": 1, "Post-COVID": 3},
    "Screen Time": {
        "Pre-COVID": [20, 20, 60],
        "Post-COVID": [10, 20, 70],
    },
    "Commute Time (Minutes)": {"Pre-COVID": 30, "Post-COVID": 90},
}

# Plot 1: Sleep Patterns
plt.figure(figsize=(6, 4))
plt.bar(data["Sleep Patterns"].keys(), data["Sleep Patterns"].values(), color=["blue", "green"])
plt.title("Sleep Patterns: Pre-COVID vs Post-COVID")
plt.ylabel("Average Hours of Sleep")
plt.savefig("sleep_patterns.png")

# Plot 2: Fast Food Consumption
plt.figure(figsize=(6, 4))
plt.bar(data["Fast Food Consumption"].keys(), data["Fast Food Consumption"].values(), color=["orange", "red"])
plt.title("Fast Food Consumption: Pre-COVID vs Post-COVID")
plt.ylabel("Relative Consumption Level")
plt.savefig("fast_food_consumption.png")

# Plot 3: Screen Time
screen_time_labels = ["Social Media", "Work", "Entertainment"]
pre_covid = data["Screen Time"]["Pre-COVID"]
post_covid = data["Screen Time"]["Post-COVID"]

fig, ax = plt.subplots(1, 2, figsize=(12, 4))
ax[0].pie(pre_covid, labels=screen_time_labels, autopct="%1.1f%%", colors=["cyan", "purple", "yellow"])
ax[0].set_title("Screen Time: Pre-COVID")

ax[1].pie(post_covid, labels=screen_time_labels, autopct="%1.1f%%", colors=["cyan", "purple", "yellow"])
ax[1].set_title("Screen Time: Post-COVID")
plt.savefig("screen_time.png")

# Plot 4: Commute Time
plt.figure(figsize=(6, 4))
plt.bar(data["Commute Time (Minutes)"].keys(), data["Commute Time (Minutes)"].values(), color=["green", "brown"])
plt.title("Commute Time: Pre-COVID vs Post-COVID")
plt.ylabel("Time (Minutes)")
plt.savefig("commute_time.png")
```
